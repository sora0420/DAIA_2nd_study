```python
import numpy as np
import pandas as pd
import sqlite3

jongmok = ['', '']

seq_length = 7 #7  

# tinker data load
con = sqlite3.connect("./kospi.db")
df_codelist = pd.read_sql("SELECT * FROM codelist", con, index_col=None)
selected = df_codelist.loc[df_codelist['name'].isin(jongmok)]

# list of dataframe
dfs = []
for n in selected['code']: #
    db_name = "SELECT * FROM " + "'" + n + "'"
    dfs.append(pd.read_sql(db_name, con, index_col=None))
    
#del dfs[0]['Date'] #0   del
#del dfs[0]['Adj Close']
#print(dfs[0])#db 
#print(make_batch(dfs[0]))        

def make_batch(source): #
    input_batch = []
    target_batch = [] #0~6  7  / 1~7  8 
#     
    del source['Adj Close']
    del source['Date']

    source = source[['Open', 'High', 'Low', 'Volume', 'Close']]
    xy = np.array(source)
    #print(xy)#source 
    #print(source) 
    y = xy[:,[-1]] #   
    #print(y) 
    #print(np.shape(y)) 861,1
    
    for i in range(len(y)-seq_length): #861 - 7 / 0~6, 1~7..
        input_batch.append(xy[i:i+seq_length])
        target_batch.append(y[i:i+seq_length])
    print(input_batch) #source
    print(target_batch) #
    print(np.shape(target_batch))

    return input_batch, target_batch
    
    
if __name__ == "__main__":
    input_batch, target_batch = make_batch(dfs[0])
```
