```python
def n_plot(_cat, n):
    d = cat.shape[1]//n # 3 
    for i in range(n):
        plt.subplot(1,n,i+1)
        plt.imshow(cat[:,i*d:(i+1)*d])
        plt.axis('off')
n  = 4
cat = data.chelsea()
n_plot(cat, n)
```

```python
n  = 3

cat = data.chelsea()
d = cat.shape[1]//3 # 3 
plt.subplot(1,3,1)
plt.imshow(cat[:,:d])
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cat[:,d:2*d])
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cat[:,2*d:3*d])
plt.axis('off')
```

```python
linear0 = np.linspace(0, 1, 2500).reshape((50,50)) #0~1 
linear1 = np.linspace(0, 255, 2500).reshape((50,50)).astype(np.uint8) #0~255 

print("linear0:", linear0.dtype, linear0.min(), linear0.max())
print("linear1:", linear1.dtype, linear1.min(), linear1.max())

fig, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(linear0, cmap='gray')
ax1.imshow(linear1, cmap='gray')
```

```python
from skimage import img_as_float, img_as_ubyte

image = data.chelsea()

image_float = img_as_float(image) #float 
image_ubyte = img_as_ubyte(image) #int

print("type, min, max:", image_float.dtype, image_float.min(), image_float.max())
print("type, min, max:", image_ubyte.dtype, image_ubyte.min(), image_ubyte.max())

print("231/255 =", 231/255.) #0~1  
```
