```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_image = np.random.random([500,500]) #랜덤함수
print(random_image)
plt.imshow(random_image, cmap='gray', interpolation='nearest') #interpolation='nearest' 2x2 matrix를 2배로 늘렸을때 나머지 빈칸들을 어떻게 채울것인가. (안쪽) / 바깥쪽이 따로 있음
```

```python
from skimage import data 

coins = data.coins() #coin의 data를 불러옴

print(coins)#위에는 random이라서 뒤죽박죽이라 노이즈처럼 보임

print(type(coins), coins.dtype, coins.shape)
plt.imshow(coins, cmap='gray', interpolation='nearest') #gray가 흑백사진
```

```python
coins = data.coins()
print(coins)
#coins[0,0] = 1pixel
#검(0) -> 흰
#흰(0) -> 검
height = len(coins) #coins.shape[0] / shape으로 접근해도 됨
width = len(coins[0]) #임의의 행을 가지고옴 / coins.shape[1]

print(height, width) #이미지 크기 shape으로 보면 (303,384)로 나옴

plt.imshow(coins, cmap='gray', interpolation='nearest')
```

```python
def n_plot(_cat, n):
    d = cat.shape[1]//n #소수를버리고 3개로 나눔
    for i in range(n):
        plt.subplot(1,n,i+1)
        plt.imshow(cat[:,i*d:(i+1)*d])
        plt.axis('off')
n  = 4
cat = data.chelsea()
n_plot(cat, n)
```

```python
n  = 3

cat = data.chelsea()
d = cat.shape[1]//3 #소수를버리고 3개로 나눔
plt.subplot(1,3,1)
plt.imshow(cat[:,:d])
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cat[:,d:2*d])
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cat[:,2*d:3*d])
plt.axis('off')
```

0 - 255 표기법: 0 은 검정색, 255 는 흰색 #int type

0 - 1 표기법: 0 은 검정색, 1 은 흰색 #float type

scikit-image는 양쪽을 모두 지원한다. 배열 자료구조에 따라 결정된다.

```python
linear0 = np.linspace(0, 1, 2500).reshape((50,50)) #0~1 표기법
linear1 = np.linspace(0, 255, 2500).reshape((50,50)).astype(np.uint8) #0~255 표기법

print("linear0:", linear0.dtype, linear0.min(), linear0.max())
print("linear1:", linear1.dtype, linear1.min(), linear1.max())

fig, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(linear0, cmap='gray')
ax1.imshow(linear1, cmap='gray')
```

```python
from skimage import img_as_float, img_as_ubyte

image = data.chelsea()

image_float = img_as_float(image) #float형식으로 캐스팅
image_ubyte = img_as_ubyte(image) #int형식으로

print("type, min, max:", image_float.dtype, image_float.min(), image_float.max())
print("type, min, max:", image_ubyte.dtype, image_ubyte.min(), image_ubyte.max())

print("231/255 =", 231/255.) #0~1 표기법으로 바꾸려고
```
